```python
%matplotlib inline
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
import camPy as cy

noisy_moons,y = datasets.make_moons(n_samples=100, noise=.1,random_state=102)
```

```python
noisy_moons[y==1,1] = noisy_moons[y==1,1]-0.7
index = np.arange(noisy_moons.shape[0])[y==0]
index = index[np.argmax(noisy_moons[y==0,0])]
#print(noisy_moons[index])
#noisy_moons[index,1] = noisy_moons[index,1]-.8
#noisy_moons[index,0] = noisy_moons[index,0]-.1
#index = np.arange(noisy_moons.shape[0])[y==0]
#index = index[np.argmax(noisy_moons[y==0,0])]
#print(noisy_moons[index])
noisy_moons[index,1] = noisy_moons[index,1]-1
noisy_moons[index,0] = noisy_moons[index,0]-.1
#noisy_moons[7,1] = noisy_moons[7,1]+0.75
nn = cy.NN_graph_fast(noisy_moons,6,metric='euclidean',use_values=True)
plt.figure(figsize=(20,8))
for i in np.arange(nn.shape[0]):
    for j in np.arange((nn[i]>0).sum()):
        index = np.array((nn[i] > 0).todense()).ravel()
        plt.plot((noisy_moons[i,0],noisy_moons[index][j,0]),(noisy_moons[i,1],noisy_moons[index][j,1]),c='grey',
                 lw=nn[i,np.arange(nn.shape[0])[index][j]])
plt.scatter(noisy_moons[0:,0],noisy_moons[0:,1],s=150)
plt.axis('off')
```

```python
l=cy.spectral_ncut(nn,2)
plt.figure(figsize=(20,12))
color = plt.cm.rainbow(l/(2-1))
for i in np.arange(nn.shape[0]):
    for j in np.arange((nn[i]>0).sum()):
        index = np.array((nn[i] > 0).todense()).ravel()
        plt.plot((noisy_moons[i,0],noisy_moons[index][j,0]),(noisy_moons[i,1],noisy_moons[index][j,1]),c='grey',
                 lw=nn[i,np.arange(nn.shape[0])[index][j]])
plt.scatter(noisy_moons[0:,0],noisy_moons[0:,1],s=200,c=color)
plt.axis('off')
```

```python
import scipy.sparse
from scipy.sparse.linalg import eigsh
import sklearn.cluster
D = scipy.sparse.diags(np.array(np.divide(1,nn.sum(axis=1))).ravel(),dtype=np.float64)
L = D*nn
w,v = eigsh(L,k=2,tol=1e-5)
clusterer = sklearn.cluster.KMeans(2)
l= clusterer.fit_predict(v[0:,0:1])
plt.scatter(np.arange(nn.shape[0]),np.sort(v[0:,1]))
```
